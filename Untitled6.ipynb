```python
# Make the colors repeat and assume
# that you don't know in advance how many unique values there
# use the len function to dynamically set the
# explode parameter based on the number of unique values.
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from itertools import cycle
# Load the dataset from a CSV file
wainaina = pd.read_csv('/content/wainaina(1).csv')
# Specify the column you want to visualize
column_name = 'Age'  # Replace with the name of your categorical column
# Calculate unique counts of the specified column
unique_counts = wainaina[column_name].value_counts()
# Determine the number of unique values
num_unique_values = len(unique_counts)
# Create a color cycle for pie chart
colors = cycle(['skyblue', 'lightcoral', 'lightgreen', 'gold', 'orange', 'pink', 'cyan', 'lavender', 'lightgray'])
# Create a pie chart to visualize the unique counts
plt.figure(figsize=(8, 8))
plt.pie(unique_counts, labels=unique_counts.index, explode=[0.1] * num_unique_values, autopct='%1.2f%%', startangle=140, colors=[next(colors) for _ in range(num_unique_values)])
plt.title(f'Unique Counts of {column_name}')
# Display the plot
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.show()
```
