```python
import matplotlib.pyplot as plt
import pandas as pd
wainaina = pd.read_csv('/content/BSCIT.csv')
Name_data = wainaina["Name"]
Age_data = wainaina["Age"]
plt.plot(Name_data, Age_data, label="Distribution",color='g')
plt.plot()
plt.xlabel("Name_data")
plt.ylabel("Age_data")
plt.title("THE MORURI FAMILY DATA")
```
